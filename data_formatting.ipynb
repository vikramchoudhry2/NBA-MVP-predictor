```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"] = players["Player"].str.replace("*","", regex = False)
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined[combined["Pts Won"]>0]
```

```python
combined
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats = combined.merge(teams, how="outer",on=["Team", "Year"])
```

```python
stats
```

```python
stats.dtypes
```

```python
stats=stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"] = stats["GB"].str.replace("nan", "0")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
